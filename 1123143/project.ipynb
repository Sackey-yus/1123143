```python
mean_Fa = dataset['fixed acidity'].mean()
mean_Va =dataset['volatile acidity'].mean()
mean_c =dataset['chlorides'].mean()
mean_Fso =dataset['free sulfur dioxide'].mean()
mean_D =dataset['density'].mean()
mean_Qua =dataset['quality'].mean()








```

```python
dataset['fixed acidity']=dataset['fixed acidity'].fillna(mean_Fa)
dataset['volatile acidity']=dataset['volatile acidity'].fillna(mean_Va)
dataset['clorides']=dataset['chlorides'].fillna(mean_c)
dataset['free sulfur dioxide']=dataset['free sulfur dioxide'].fillna(mean_Fso)
dataset['density']=dataset['density'].fillna(mean_D)
dataset['quality']=dataset['quality'].fillna(mean_Qua)




```

```python

seaborn importas sns 
import matplotlib.pyplot as plt


numerical_columns = dataset.select_dtypes(include ='number').columns
num_columns = len(numerical_columns)
row = (num_columns // 3) + (num_columns % 3)
cols = 3 
plt.figure(figsize=(15,10))
for i,column in enumerate(numerical_columns,1):
    plt.subplot(row,cols,i)
    sns.violinplot(x=column, data=dataset, color='green')
    plt.title(f'violin plot of {column}')
plt.tight_layout()
plt.show()
```
